**Theoritical Questions**

1.	What is unsupervised learning in the context of machine learning?
Ans.
Unsupervised learning is a type of machine learning where the algorithm learns from unlabeled data. It aims to identify patterns, structures, or groupings within the data without prior knowledge of the outcome.   

2.	How does K-Means clustering algorithm work?
Ans.
The K-Means clustering algorithm works by partitioning data into k clusters, where each data point belongs to the cluster with the nearest mean (centroid). It iteratively assigns data points to the nearest centroid and updates the centroids based on the new cluster assignments until convergence is reached.   

3.	Explain the concept of a dendrogram in hierarchical clustering.
Ans.
A dendrogram in hierarchical clustering is a tree-like diagram that illustrates the sequence of merges or splits of clusters. The vertical axis represents the distance or similarity between clusters, and the leaves of the tree represent individual data points.   

4.	What is the main difference between K-Means and Hierarchical Clustering?
Ans.
The main difference between K-Means and Hierarchical Clustering is that K-Means partitions data into a predetermined number of non-overlapping clusters, while Hierarchical Clustering creates a hierarchy of clusters. K-Means requires the number of clusters to be specified in advance, whereas Hierarchical Clustering can produce a dendrogram that shows clusters at various levels of granularity.   

5.	What are the advantages of DBSCAN over K-Means?
Ans.
DBSCAN has advantages over K-Means in that it can discover clusters of arbitrary shapes and is robust to outliers. It does not require specifying the number of clusters beforehand and can identify noise points.   

6.	When would you use Silhouette Score in clustering?
Ans.
Silhouette Score is used to evaluate the quality of clustering results. It helps to measure how well each data point fits within its assigned cluster compared to other clusters.   

7.	What are the limitations of Hierarchical Clustering?
Ans.
Hierarchical Clustering can be computationally expensive, especially for large datasets. It can also be sensitive to noise and outliers.   

8.	Why is feature scaling important in clustering algorithms like K-Means?
Ans.
Feature scaling is important in clustering algorithms like K-Means because K-Means uses distance measures to assign data points to clusters. If features are on different scales, features with larger scales can dominate the distance calculations and skew the clustering results.   

9.	How does DBSCAN identify noise points?
Ans.
DBSCAN identifies noise points as data points that have too few neighbors within a specified radius (epsilon). These points are not part of any dense region and are considered outliers.   

10.	Define inertia in the context of K-Means.
Ans.
Inertia in the context of K-Means is the sum of the squared distances of each data point to its closest centroid. It measures the within-cluster dispersion, and lower inertia indicates tighter clusters.   

11.	What is the elbow method in K-Means clustering?
Ans.
The elbow method in K-Means clustering is a technique used to determine the optimal number of clusters (k). It involves plotting the inertia for different values of k and selecting the k at the "elbow" point of the curve, where the rate of decrease in inertia sharply changes.   

12.	Describe the concept of "density" in DBSCAN.
Ans.
In DBSCAN, "density" refers to the number of data points within a specified radius (epsilon) of a given point. Dense regions are areas with a high concentration of data points, and DBSCAN groups these regions into clusters.   

13.	Can hierarchical clustering be used on categorical data?
Ans.
The document does not explicitly state whether hierarchical clustering can be used on categorical data.

14.	What does a negative Silhouette Score indicate?
Ans.
A negative Silhouette Score indicates that a data point might be assigned to the wrong cluster. It suggests that the data point is, on average, closer to the data points in a neighboring cluster than to the data points in its own cluster.   

15.	Explain the term "linkage criteria" in hierarchical clustering.
Ans. Linkage criteria in hierarchical clustering determine the distance between clusters. Different linkage methods (e.g., single, complete, average) define how this distance is calculated, influencing the shape and separation of the resulting clusters.   

16.	Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?
Ans.
K-Means clustering might perform poorly on data with varying cluster sizes or densities because it assumes that clusters are spherical and of similar size. It tends to split large clusters and merge small or dense clusters.   

17.	What are the core parameters in DBSCAN, and how do they influence clustering?
Ans.
The core parameters in DBSCAN are epsilon (eps) and minPts. Epsilon defines the radius to search for neighbors, and minPts is the minimum number of points required to form a dense region. A small epsilon can lead to many noise points, while a large epsilon can merge clusters. A small minPts can result in noise being included in clusters, and a large minPts can lead to the formation of small, fragmented clusters.   

18.	How does K-Means++ improve upon standard K-Means initialization?
Ans.
K-Means++ improves upon standard K-Means initialization by selecting initial centroids that are well-separated from each other. This initialization technique helps to speed up convergence and improve the quality of the resulting clusters.   

19.	What is agglomerative clustering?
Ans.
Agglomerative clustering is a hierarchical clustering approach that starts with each data point as its own cluster and iteratively merges the closest clusters until a single cluster remains.   

20.	What makes Silhouette Score a better metric than just inertia for model evaluation?
Ans.
Silhouette Score is a better metric than just inertia for model evaluation because it considers both the cohesion within clusters and the separation between clusters. Inertia only measures within-cluster dispersion, while Silhouette Score provides a more balanced assessment of clustering quality.   



```python
# PRACTICAL QUESTIONS
''' Question 21.Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a
scatter plot.'''
#Ans.
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with 4 centers
# n_samples: The total number of points equally divided among clusters.
# centers: The number of centers to generate.
# random_state: Determines random number generation for dataset creation. Pass an int for reproducible output across multiple function calls.
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Apply K-Means clustering with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)  # Initialize KMeans
kmeans.fit(X)  # Fit the model to the data
labels = kmeans.labels_  # Get the cluster labels
centers = kmeans.cluster_centers_ # Get the cluster centers

# Visualize the data and the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8)  # Plot the data points with color-coded labels
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='o', s=200, label='Cluster Centers')  # Plot the cluster centers
plt.title('K-Means Clustering with 4 Clusters')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
'''Question 23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot'''
#Answer
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Generate synthetic data using make_moons
X, y = make_moons(n_samples=300, noise=0.15, random_state=42) # Increased noise for more outliers

# Apply DBSCAN
# eps: The maximum distance between two samples for one to be considered as in the neighborhood of the other.
# min_samples: The number of samples (or total weight) in a neighborhood for a point to be considered as a core point.
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(X)
labels = dbscan.labels_

# Identify outliers (noise points)
outliers = X[labels == -1]

# Visualize the data and the clusters
plt.figure(figsize=(8, 6))
# Plot the clustered points.
plt.scatter(X[labels != -1, 0], X[labels != -1, 1], c=labels[labels != -1], cmap='viridis', s=50, alpha=0.8)
# Plot the outliers
plt.scatter(outliers[:, 0], outliers[:, 1], c='red', marker='o', s=100, label='Outliers')
plt.title('DBSCAN Clustering with make_moons')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
'''Question 24.  Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each
cluster'''
#Answer:
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import numpy as np

# Load the Wine dataset
wine = load_wine()
X = wine.data

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
labels = kmeans.labels_

# Print the size of each cluster
cluster_sizes = np.bincount(labels)
for i, size in enumerate(cluster_sizes):
    print(f"Cluster {i}: {size} samples")
```

```python
'''Question 29.  Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart'''
#Answer.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Calculate silhouette scores for k = 2 to 5
silhouette_scores = []
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    labels = kmeans.labels_
    silhouette_avg = silhouette_score(X, labels)
    silhouette_scores.append(silhouette_avg)

# Display the silhouette scores in a bar chart
plt.figure(figsize=(8, 6))
plt.bar(range(2, 6), silhouette_scores)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Values of k')
plt.xticks(range(2, 6))
plt.grid(True)
plt.show()
```

```python
'''Question 30.Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage'''
#Answer
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

# Load the Iris dataset
iris = load_iris()
X = iris.data

# Apply Hierarchical Clustering with average linkage
linked = linkage(X, 'average')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Sample Index')
plt.ylabel('Cluster Distance')
plt.show()
```

```python
'''Question 31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with
decision boundaries'''
#Answer. import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Generate synthetic data with overlapping clusters
X, y = make_blobs(n_samples=300, centers=3, cluster_std=2.0, random_state=42)

# Apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

# Define a grid of points to plot the decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Visualize the data, cluster assignments, and decision boundaries
plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3)  # Plot the decision boundaries
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8) # Plot the data points
plt.title('K-Means Clustering with Overlapping Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

```python
'''Question 33.  Generate synthetic data using make_blobs and apply Agglomerative Clustering with complete linkage. Plot
the result'''

#Answer.
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
import numpy as np

# Generate synthetic data using make_blobs
# n_samples: The total number of points equally divided among clusters.
# centers: The number of centers to generate, or the fixed center locations.
# n_features: The number of features for each sample.
# random_state: Determines random number generation for dataset creation. Pass an int for reproducible output across multiple function calls.
X, y = make_blobs(n_samples=300, centers=4, n_features=2, random_state=42)

# Apply Agglomerative Clustering with complete linkage
# n_clusters: The number of clusters to find.
# linkage: {“ward”, “complete”, “average”, “single”}, which linkage criterion to use.
#     - complete: maximum distance between all observations of the two sets.
agg_clustering = AgglomerativeClustering(n_clusters=4, linkage='complete')
y_pred = agg_clustering.fit_predict(X)

# Plot the results
plt.figure(figsize=(8, 6))
# Use a colormap ('viridis') for better visual distinction between clusters.
# Use np.unique to determine the number of unique labels dynamically.
unique_labels = np.unique(y_pred)
for label in unique_labels:
    # Plot data points, using a marker size of 10.
    plt.scatter(X[y_pred == label, 0], X[y_pred == label, 1], label=f'Cluster {label}', s=30)
plt.title('Agglomerative Clustering with Complete Linkage')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True) # Add a grid for better readability.
plt.show()
```

```python
'''Question 34. Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a
line plot'''
#Answer.

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import KMeans

# Load the Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data

# Calculate inertia for K = 2 to 6
inertia_values = []
for k in range(2, 7):
    # Initialize KMeans with the current value of k
    kmeans = KMeans(n_clusters=k, random_state=42)
    # Fit KMeans to the data
    kmeans.fit(X)
    # Append the inertia value to the list
    inertia_values.append(kmeans.inertia_)

# Plot the results
plt.figure(figsize=(8, 6))
# Plot the inertia values against the number of clusters (K)
plt.plot(range(2, 7), inertia_values, marker='o', linestyle='-', color='b')
plt.title('K-Means Inertia for Breast Cancer Dataset')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(range(2, 7)) # Ensure that all K values are shown on the x-axis
plt.grid(True) # Add grid lines for better readability
plt.show()
```

```python
'''Question 40. Load the Digits dataset, apply PCA to reduce to 3 components, then use KMeans and visualize with a 3D
scatter plot'''
#Answer.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import numpy as np
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting

# Load the Digits dataset
digits = load_digits()
X = digits.data
y = digits.target # Store the target labels for coloring the plot

# Apply PCA to reduce to 3 components
pca = PCA(n_components=3, random_state=42)
X_pca = pca.fit_transform(X)

# Apply KMeans
# n_clusters: The number of clusters to form as well as the number of centroids to generate.
# random_state: Determines random number generation for centroid initialization. Use an int for reproducibility.
kmeans = KMeans(n_clusters=10, random_state=42)  # Digits dataset has 10 classes
kmeans.fit(X_pca)
y_pred = kmeans.predict(X_pca)
centers = kmeans.cluster_centers_

# Visualize the results with a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')  # Create a 3D subplot

# Use the target labels (y) to color the data points, for better visualization
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, cmap='viridis', s=30) # Color by true class

# Plot the cluster centers as red crosses
ax.scatter(centers[:, 0], centers[:, 1], centers[:, 2], c='red', marker='X', s=200, label='Cluster Centers')

ax.set_title('K-Means Clustering of Digits Dataset with PCA (3D)')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.legend()

# Add a colorbar to indicate the digit labels
colorbar = fig.colorbar(scatter, ax=ax)
colorbar.set_label('Digit Class')

plt.show()

```

```python
'''Question 41. Generate synthetic blobs with 5 centers and apply KMeans. Then use silhouette_score to evaluate the
clustering'''
#Answer.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Generate synthetic data using make_blobs
# n_samples: The total number of points equally divided among clusters.
# centers: The number of centers to generate, or the fixed center locations.
# n_features: The number of features for each sample.
# random_state: Determines random number generation for dataset creation.
X, y = make_blobs(n_samples=300, centers=5, n_features=2, random_state=42)

# Apply KMeans
# n_clusters: The number of clusters to form as well as the number of centroids to generate.
# random_state: Determines random number generation for centroid initialization. Use an int for reproducibility.
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)
y_pred = kmeans.predict(X)

# Evaluate the clustering using silhouette_score
# X: The feature matrix.
# labels: The predicted cluster labels.
silhouette_avg = silhouette_score(X, y_pred)
print(f"Silhouette Score: {silhouette_avg}")

# Plot the results
plt.figure(figsize=(8, 6))
# Use a colormap ('viridis') for better visual distinction between clusters.
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis', s=30)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

```python
'''Question 42. Load the Breast Cancer dataset, reduce dimensionality using PCA, and apply Agglomerative Clustering.
Visualize in 2D '''
#Answer.

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
import numpy as np

# Load the Breast Cancer dataset
cancer = load_breast_cancer()
X = cancer.data

# Apply PCA to reduce to 2 components
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

# Apply Agglomerative Clustering
# n_clusters: The number of clusters to find.
# linkage: {“ward”, “complete”, “average”, “single”}, which linkage criterion to use.
agg_clustering = AgglomerativeClustering(n_clusters=2, linkage='ward')
y_pred = agg_clustering.fit_predict(X_pca)

# Plot the results
plt.figure(figsize=(8, 6))
# Use a colormap ('viridis') for better visual distinction between clusters.
unique_labels = np.unique(y_pred) # Get unique labels to iterate over.
for label in unique_labels:
    # Plot data points, using a marker size of 30.
    plt.scatter(X_pca[y_pred == label, 0], X_pca[y_pred == label, 1], label=f'Cluster {label}', s=30)
plt.title('Agglomerative Clustering on Breast Cancer Dataset with PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
''' Question 43.Generate noisy circular data using make_circles and visualize clustering results from KMeans and DBSCAN
side-by-side '''
#Answer.

import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
from sklearn.cluster import KMeans, DBSCAN

# Generate noisy circular data using make_circles
# n_samples: The total number of points generated.
# factor: The ratio of the inner circle radius to the outer circle radius.
# noise: Standard deviation of Gaussian noise added to the data.
# random_state: Determines random number generation for dataset creation.
X, y = make_circles(n_samples=300, factor=0.5, noise=0.1, random_state=42)

# Apply KMeans
# n_clusters: The number of clusters to form as well as the number of centroids to generate.
# random_state: Determines random number generation for centroid initialization. Use an int for reproducibility.
kmeans = KMeans(n_clusters=2, random_state=42)
y_pred_kmeans = kmeans.fit_predict(X)

# Apply DBSCAN
# eps: The maximum distance between two samples for one to be considered as in the neighborhood of the other.
# min_samples: The number of samples (or total weight) in a neighborhood for a point to be considered as a core point.
dbscan = DBSCAN(eps=0.2, min_samples=5)  # Adjust eps as needed for the dataset
y_pred_dbscan = dbscan.fit_predict(X)

# Visualize the results side-by-side
plt.figure(figsize=(12, 6))

# Plot KMeans results
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y_pred_kmeans, cmap='viridis', s=30)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

# Plot DBSCAN results
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=y_pred_dbscan, cmap='viridis', s=30)
plt.title('DBSCAN Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)

plt.tight_layout()  # Adjust layout to prevent overlapping titles
plt.show()
```

```python
''' Question 47.  Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the
count '''
#Answer

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

# Generate synthetic data with make_blobs
# n_samples: The total number of points equally divided among clusters.
# centers: The number of centers to generate, or the fixed center locations.
# n_features: The number of features for each sample.
# random_state: Determines random number generation for dataset creation.
X, y = make_blobs(n_samples=300, centers=3, n_features=2, random_state=42)

# Inject some noise by adding random points
noise = np.random.rand(30, 2)  # 10% noise
X = np.concatenate((X, noise), axis=0)

# Apply DBSCAN
# eps: The maximum distance between two samples for one to be considered as in the neighborhood of the other.
# min_samples: The number of samples (or total weight) in a neighborhood for a point to be considered as a core point.
dbscan = DBSCAN(eps=0.8, min_samples=5)
dbscan.fit(X)
y_pred = dbscan.labels_

# Count the number of clusters and noise points
n_clusters = len(set(y_pred)) - (1 if -1 in y_pred else 0)
n_noise_points = list(y_pred).count(-1)

# Print the results
print(f"Number of clusters: {n_clusters}")
print(f"Number of noise points: {n_noise_points}")

# Visualize the results
plt.figure(figsize=(8, 6))
# Use a colormap ('viridis') for better visual distinction between clusters.
unique_labels = np.unique(y_pred)
for label in unique_labels:
    # Plot data points, using a marker size of 30.
    if label == -1:
        plt.scatter(X[y_pred == label, 0], X[y_pred == label, 1], label='Noise', s=30, alpha=0.5, color='gray')
    else:
        plt.scatter(X[y_pred == label, 0], X[y_pred == label, 1], label=f'Cluster {label}', s=30)

plt.title('DBSCAN Clustering on Noisy Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
''' Question 48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the
clusters'''
#Answer

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.cluster import AgglomerativeClustering
from sklearn.manifold import TSNE
import numpy as np

# Load the Digits dataset
digits = load_digits()
X = digits.data

# Apply t-SNE to reduce to 2 dimensions
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Apply Agglomerative Clustering
# n_clusters: The number of clusters to find.
# linkage: {“ward”, “complete”, “average”, “single”}, which linkage criterion to use.
agg_clustering = AgglomerativeClustering(n_clusters=10, linkage='ward')  # Use 10 clusters for digits
y_pred = agg_clustering.fit_predict(X_tsne)

# Plot the results
plt.figure(figsize=(8, 6))
# Use a colormap ('viridis') for better visual distinction between clusters.
unique_labels = np.unique(y_pred)
for label in unique_labels:
    # Plot data points, using a marker size of 30.
    plt.scatter(X_tsne[y_pred == label, 0], X_tsne[y_pred == label, 1], label=f'Cluster {label}', s=30)
plt.title('Agglomerative Clustering on Digits Dataset with t-SNE')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.legend()
plt.grid(True)
plt.show()
```
