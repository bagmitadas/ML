Q1. What is hypothesis testing in statistics?
Ans. Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.   

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
Ans. The null hypothesis (H0) is a statement of no effect or no difference, while the alternative hypothesis (H1 or Ha) is a statement that contradicts the null hypothesis, suggesting there is a significant effect or difference.   

Q3. What is the significance level in hypothesis testing, and why is it important?
Ans. The significance level (alpha or α) is the probability of rejecting the null hypothesis when it is actually true. It is important because it sets the threshold for determining statistical significance.   

Q4. What does a P-value represent in hypothesis testing?
Ans. A P-value represents the probability of obtaining the observed results (or more extreme results) if the null hypothesis is true.   

Q5. How do you interpret the P-value in hypothesis testing?
Ans. If the P-value is less than or equal to the significance level (α), you reject the null hypothesis; otherwise, you fail to reject the null hypothesis.   

Q6. What are Type 1 and Type 2 errors in hypothesis testing?
Ans. A Type 1 error occurs when you reject the null hypothesis when it is true, while a Type 2 error occurs when you fail to reject the null hypothesis when it is false.   

Q7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
Ans. A one-tailed test examines if the sample mean is significantly greater than or less than the population mean, while a two-tailed test examines if the sample mean is significantly different from the population mean (in either direction).   

Q8. What is the Z-test, and when is it used in hypothesis testing?
Ans. A Z-test is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large.   

Q9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
Ans. The Z-score is calculated as (sample mean - population mean) / (population standard deviation / square root of sample size). It represents how many standard deviations a data point is from the mean.   

Q10. What is the T-distribution, and when should it be used instead of the normal distribution?
Ans. The T-distribution is a probability distribution used when the sample size is small or the population standard deviation is unknown.   

Q11. What is the difference between a Z-test and a T-test?
Ans. A Z-test is used when the population standard deviation is known or the sample size is large, while a T-test is used when the population standard deviation is unknown or the sample size is small.   

Q12. What is the T-test, and how is it used in hypothesis testing?
Ans. A T-test is a statistical test used to compare the means of two groups.   

Q13. What is the relationship between Z-test and T-test in hypothesis testing?
Ans. The T-test is similar to the Z-test, but it is used when the population standard deviation is unknown or the sample size is small. As the sample size increases, the T-distribution approaches the Z-distribution.   

Q14. What is a confidence interval, and how is it used to interpret statistical results?
Ans. A confidence interval is a range of values that is likely to contain the true population parameter. It is used to estimate the precision of a sample statistic.   

Q15. What is the margin of error, and how does it affect the confidence interval?
Ans. The margin of error is the range above and below the sample statistic in a confidence interval. A larger margin of error results in a wider confidence interval, indicating less precision.   

Q16. How is Bayes' Theorem used in statistics, and what is its significance?
Ans. Bayes' Theorem is used to update the probability of a hypothesis based on new evidence. It is significant in Bayesian inference for combining prior beliefs with observed data.   

Q17. What is the Chi-square distribution, and when is it used?
Ans. The Chi-square distribution is a probability distribution used for categorical data. It is used in tests of independence and goodness of fit.   

Q18. What is the Chi-square goodness of fit test, and how is it applied?
Ans. The Chi-square goodness of fit test is used to determine if the observed frequency distribution of a categorical variable differs from the expected distribution.   

Q19. What is the F-distribution, and when is it used in hypothesis testing?
Ans. The F-distribution is a probability distribution used in ANOVA tests to compare variances between groups.   

Q20. What is an ANOVA test, and what are its assumptions?
Ans. ANOVA (Analysis of Variance) is a statistical test used to compare the means of two or more groups. Assumptions include normality, independence, and equal variance.   

Q21. What are the different types of ANOVA tests?
Ans. The different types of ANOVA tests include one-way ANOVA, two-way ANOVA, and repeated measures ANOVA.   

Q22. What is the F-test, and how does it relate to hypothesis testing?
Ans. The F-test is a statistical test used to compare variances. In ANOVA, it is used to determine if there is a significant difference between group means.


```python
# PRACTICAL QUESTIONS

'''Q1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results'''
import numpy as np
from scipy import stats

def one_sample_z_test(sample_mean, sample_std, sample_size, population_mean, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test and interprets the results.

    Args:
        sample_mean (float): The mean of the sample.
        sample_std (float): The standard deviation of the sample.
        sample_size (int): The size of the sample.
        population_mean (float): The known mean of the population.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The alternative hypothesis ('two-sided', 'less', 'greater').
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the z-statistic, p-value, and interpretation.
    """
    standard_error = sample_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error
    p_value = stats.norm.sf(np.abs(z_statistic))  # Survival function (1 - CDF) for right tail

    if alternative == 'two-sided':
        p_value *= 2
        if p_value < alpha:
            interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                             f"We reject the null hypothesis. There is statistically significant evidence " \
                             f"that the sample mean is different from the population mean."
        else:
            interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                             f"We fail to reject the null hypothesis. There is not enough statistically " \
                             f"significant evidence that the sample mean is different from the population mean."
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
        if p_value < alpha:
            interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                             f"We reject the null hypothesis. There is statistically significant evidence " \
                             f"that the sample mean is less than the population mean."
        else:
            interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                             f"We fail to reject the null hypothesis. There is not enough statistically " \
                             f"significant evidence that the sample mean is less than the population mean."
    elif alternative == 'greater':
        if p_value < alpha:
            interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                             f"We reject the null hypothesis. There is statistically significant evidence " \
                             f"that the sample mean is greater than the population mean."
        else:
            interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                             f"We fail to reject the null hypothesis. There is not enough statistically " \
                             f"significant evidence that the sample mean is greater than the population mean."
    else:
        raise ValueError("Invalid alternative hypothesis specified. Choose 'two-sided', 'less', or 'greater'.")

    return z_statistic, p_value, interpretation

# Example usage:
sample_mean = 105
sample_std = 15
sample_size = 50
population_mean = 100
alpha = 0.05
alternative = 'two-sided'  # You can change this to 'less' or 'greater'

z_stat, p_val, result_interpretation = one_sample_z_test(sample_mean, sample_std, sample_size, population_mean, alpha, alternative)

print(f"Z-Statistic: {z_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(result_interpretation)
```

```python
''' Q2.  Simulate random data to perform hypothesis testing and calculate the corresponding P-value using Python '''
import numpy as np
from scipy import stats

def simulate_hypothesis_test(population_mean, population_std, sample_size, null_hypothesis_mean, alpha=0.05, num_simulations=1000):
    """
    Simulates random data to perform hypothesis testing and calculates the P-value.

    Args:
        population_mean (float): The true mean of the population from which data is sampled.
        population_std (float): The true standard deviation of the population.
        sample_size (int): The size of each simulated sample.
        null_hypothesis_mean (float): The mean stated in the null hypothesis.
        alpha (float, optional): The significance level. Defaults to 0.05.
        num_simulations (int, optional): The number of simulations to perform. Defaults to 1000.

    Returns:
        float: The proportion of simulations where the null hypothesis would be rejected (approximate P-value).
    """
    reject_count = 0
    for _ in range(num_simulations):
        sample_data = np.random.normal(population_mean, population_std, sample_size)
        sample_mean = np.mean(sample_data)
        standard_error = population_std / np.sqrt(sample_size)
        z_statistic = (sample_mean - null_hypothesis_mean) / standard_error
        p_value = 2 * stats.norm.sf(np.abs(z_statistic))  # Two-tailed test

        if p_value < alpha:
            reject_count += 1

    empirical_p_value = reject_count / num_simulations
    return empirical_p_value

# Example usage:
population_mean = 100
population_std = 15
sample_size = 30
null_hypothesis_mean = 102
alpha = 0.05
num_simulations = 1000

p_value = simulate_hypothesis_test(population_mean, population_std, sample_size, null_hypothesis_mean, alpha, num_simulations)
print(f"Approximate P-value from simulation: {p_value:.3f}")
```

```python
 ''' Q3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean '''
 import numpy as np
from scipy import stats

def one_sample_z_test_direct(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test using sample data.

    Args:
        sample_data (array-like): The data from the sample.
        population_mean (float): The known mean of the population.
        population_std (float): The known standard deviation of the population.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The alternative hypothesis ('two-sided', 'less', 'greater').
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the z-statistic and p-value.
    """
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error
    p_value = stats.norm.sf(np.abs(z_statistic))

    if alternative == 'two-sided':
        p_value *= 2
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        pass  # p_value is already correctly calculated for 'greater'

    return z_statistic, p_value

# Example usage:
np.random.seed(42)
population_mean = 50
population_std = 10
sample_data = np.random.normal(48, 10, 40)
alpha = 0.05
alternative = 'two-sided'

z_stat, p_val = one_sample_z_test_direct(sample_data, population_mean, population_std, alpha, alternative)

print(f"Z-Statistic: {z_stat:.3f}")
print(f"P-Value: {p_val:.3f}")

if p_val < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Q4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test_and_plot(sample_mean, sample_std, sample_size, population_mean, alpha=0.05):
    """
    Performs a two-tailed Z-test and visualizes the decision region.

    Args:
        sample_mean (float): The mean of the sample.
        sample_std (float): The standard deviation of the sample.
        sample_size (int): The size of the sample.
        population_mean (float): The known mean of the population.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the z-statistic and p-value.
    """
    standard_error = sample_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error
    p_value = 2 * stats.norm.sf(np.abs(z_statistic))

    # Visualize the decision region
    x = np.linspace(population_mean - 4 * standard_error, population_mean + 4 * standard_error, 200)
    plt.plot(x, stats.norm.pdf(x, population_mean, standard_error), label='Sampling Distribution of Mean')
    plt.axvline(population_mean, color='k', linestyle='dashed', linewidth=1, label='Population Mean')

    critical_value = stats.norm.ppf(1 - alpha/2)
    plt.axvline(population_mean - critical_value * standard_error, color='r', linestyle='dashed', linewidth=1, label=f'Critical Values (±{critical_value:.2f} SE)')
    plt.axvline(population_mean + critical_value * standard_error, color='r', linestyle='dashed', linewidth=1)

    plt.axvline(sample_mean, color='g', linestyle='solid', linewidth=1, label=f'Sample Mean ({sample_mean:.2f})')

    plt.title('Two-Tailed Z-Test and Decision Region')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_statistic, p_value

# Example usage:
sample_mean = 52
sample_std = 10
sample_size = 100
population_mean = 50
alpha = 0.05

z_stat, p_val = two_tailed_z_test_and_plot(sample_mean, sample_std, sample_size, population_mean, alpha)

print(f"Z-Statistic: {z_stat:.3f}")
print(f"P-Value: {p_val:.3f}")

if p_val < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# Q5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(null_mean, alt_mean, std_dev, sample_size, alpha=0.05):
    """
    Calculates and visualizes Type 1 and Type 2 errors.

    Args:
        null_mean (float): The mean under the null hypothesis.
        alt_mean (float): The mean under the alternative hypothesis.
        std_dev (float): The population standard deviation (assumed known for Z-test).
        sample_size (int): The sample size.
        alpha (float, optional): The significance level (probability of Type 1 error). Defaults to 0.05.
    """
    standard_error = std_dev / np.sqrt(sample_size)
    critical_value_right = null_mean + stats.norm.ppf(1 - alpha) * standard_error
    critical_value_left = null_mean - stats.norm.ppf(1 - alpha) * standard_error # For two-tailed

    # For visualization
    x = np.linspace(null_mean - 4 * standard_error, alt_mean + 4 * standard_error, 200)
    y_null = stats.norm.pdf(x, null_mean, standard_error)
    y_alt = stats.norm.pdf(x, alt_mean, standard_error)

    plt.plot(x, y_null, label=f'Null Hypothesis (Mean={null_mean:.2f})')
    plt.plot(x, y_alt, label=f'Alternative Hypothesis (Mean={alt_mean:.2f})')

    # Shade Type 1 error region (assuming two-tailed for visualization)
    x_crit_left = np.linspace(null_mean - 4 * standard_error, critical_value_left, 100)
    plt.fill_between(x_crit_left, 0, stats.norm.pdf(x_crit_left, null_mean, standard_error), color='red', alpha=0.3, label='Type 1 Error (α/2)')
    x_crit_right = np.linspace(critical_value_right, null_mean + 4 * standard_error, 100)
    plt.fill_between(x_crit_right, 0, stats.norm.pdf(x_crit_right, null_mean, standard_error), color='red', alpha=0.3)

    # Shade Type 2 error region
    x_beta = np.linspace(null_mean - 4 * standard_error, critical_value_right, 100) # Using right critical value for illustration
    plt.fill_between(x_beta, 0, stats.norm.pdf(x_beta, alt_mean, standard_error), color='blue', alpha=0.3, label='Type 2 Error (β)')

    plt.axvline(critical_value_left, color='k', linestyle='--', linewidth=1, label='Critical Values')
    plt.axvline(critical_value_right, color='k', linestyle='--', linewidth=1)

    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.title('Visualization of Type 1 and Type 2 Errors')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Calculate Type 2 error (beta) for the given alternative mean
    beta = stats.norm.cdf(critical_value_right, alt_mean, standard_error) - stats.norm.cdf(critical_value_left, alt_mean, standard_error) # For two-tailed
    power = 1 - beta
    print(f"Significance Level (Alpha): {alpha:.3f}")
    print(f"Probability of Type 2 Error (Beta): {beta:.3f}")
    print(f"Power of the Test (1 - Beta): {power:.3f}")

# Example usage:
null_mean = 50
alt_mean = 53
std_dev = 10
sample_size = 30
alpha = 0.05

visualize_type1_type2_errors(null_mean, alt_mean, std_dev, sample_size, alpha)
```

```python
# Q6. Write a Python program to perform an independent T-test and interpret the results
import numpy as np
from scipy import stats

def independent_t_test(sample1, sample2, alpha=0.05, equal_variance=True):
    """
    Performs an independent T-test and interprets the results.

    Args:
        sample1 (array-like): The data from the first sample.
        sample2 (array-like): The data from the second sample.
        alpha (float, optional): The significance level. Defaults to 0.05.
        equal_variance (bool, optional): Whether to assume equal variances between the two groups.
                                         Defaults to True (Welch's t-test is used if False).

    Returns:
        tuple: A tuple containing the t-statistic, p-value, and interpretation.
    """
    if equal_variance:
        t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=True)
    else:
        t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. There is statistically significant evidence " \
                         f"that the means of the two groups are different."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence that the means of the two groups are different."

    return t_statistic, p_value, interpretation

# Example usage:
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 12, 35)
alpha = 0.05
equal_variance = True  # You can set this to False if variances are not assumed equal

t_stat, p_val, result_interpretation = independent_t_test(group1, group2, alpha, equal_variance)

print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(result_interpretation)
```

```python
#Q 7. Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd

def paired_t_test_and_visualize(before, after, alpha=0.05):
    """
    Performs a paired sample T-test and visualizes the comparison results.

    Args:
        before (array-like): The data before the treatment/intervention.
        after (array-like): The data after the treatment/intervention.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the t-statistic, p-value, and interpretation.
    """
    difference = after - before
    t_statistic, p_value = stats.ttest_rel(before, after)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. There is statistically significant evidence " \
                         f"of a difference between the before and after measurements."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence of a difference between the before and after measurements."

    # Visualization
    df = pd.DataFrame({'Before': before, 'After': after})
    df['Difference'] = difference

    plt.figure(figsize=(8, 6))
    plt.subplot(1, 2, 1)
    plt.scatter(df.index, df['Before'], label='Before', alpha=0.7)
    plt.scatter(df.index, df['After'], label='After', alpha=0.7)
    for i in df.index:
        plt.plot([i, i], [df['Before'][i], df['After'][i]], linestyle='--', color='gray')
    plt.xlabel('Observation Index')
    plt.ylabel('Value')
    plt.title('Before vs. After Measurements')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.hist(df['Difference'], bins='auto', color='skyblue', edgecolor='black')
    plt.axvline(df['Difference'].mean(), color='red', linestyle='dashed', linewidth=1, label=f'Mean Difference: {df["Difference"].mean():.2f}')
    plt.xlabel('Difference (After - Before)')
    plt.ylabel('Frequency')
    plt.title('Distribution of Differences')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

    return t_statistic, p_value, interpretation

# Example usage:
np.random.seed(42)
before_treatment = np.random.normal(60, 15, 40)
after_treatment = before_treatment + np.random.normal(5, 10, 40) # Simulate some effect
alpha = 0.05

t_stat, p_val, result_interpretation = paired_t_test_and_visualize(before_treatment, after_treatment, alpha)

print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(result_interpretation)
```

```python
# Q8. Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
from scipy import stats

def compare_z_and_t_tests(population_mean, population_std, sample_size, alpha=0.05, num_simulations=1000):
    """
    Simulates data, performs Z-test and T-test, and compares the rejection rates.

    Args:
        population_mean (float): The true population mean.
        population_std (float): The true population standard deviation.
        sample_size (int): The size of each simulated sample.
        alpha (float, optional): The significance level. Defaults to 0.05.
        num_simulations (int, optional): The number of simulations to perform. Defaults to 1000.

    Returns:
        tuple: A tuple containing the rejection rates for Z-test and T-test.
    """
    z_reject_count = 0
    t_reject_count = 0

    for _ in range(num_simulations):
        sample_data = np.random.normal(population_mean, population_std, sample_size)
        sample_mean = np.mean(sample_data)
        sample_std_est = np.std(sample_data, ddof=1) # Sample standard deviation (for T-test)
        standard_error_z = population_std / np.sqrt(sample_size)
        standard_error_t = sample_std_est / np.sqrt(sample_size)

        # Z-test
        z_statistic = (sample_mean - population_mean) / standard_error_z
        p_value_z = 2 * stats.norm.sf(np.abs(z_statistic))
        if p_value_z < alpha:
            z_reject_count += 1

        # T-test
        t_statistic = (sample_mean - population_mean) / standard_error_t
        degrees_of_freedom = sample_size - 1
        p_value_t = 2 * stats.t.sf(np.abs(t_statistic), df=degrees_of_freedom)
        if p_value_t < alpha:
            t_reject_count += 1

    z_rejection_rate = z_reject_count / num_simulations
    t_rejection_rate = t_reject_count / num_simulations

    print(f"Simulated rejection rate (Z-test): {z_rejection_rate:.3f}")
    print(f"Simulated rejection rate (T-test): {t_rejection_rate:.3f}")
    print("\nComparison:")
    if abs(z_rejection_rate - alpha) < abs(t_rejection_rate - alpha):
        print("The Z-test's rejection rate is closer to the significance level (alpha).")
    elif abs(t_rejection_rate - alpha) < abs(z_rejection_rate - alpha):
        print("The T-test's rejection rate is closer to the significance level (alpha).")
    else:
        print("Both tests have similar rejection rates close to the significance level (alpha).")
    print("Note: When the population standard deviation is known, the Z-test is theoretically more accurate. However, in practice with large enough sample sizes, the T-test often yields similar results.")

    return z_rejection_rate, t_rejection_rate

# Example usage:
population_mean = 50
population_std = 10
sample_size = 30
alpha = 0.05
num_simulations = 1000

z_rate, t_rate = compare_z_and_t_tests(population_mean, population_std, sample_size, alpha, num_simulations)
```

```python
# Q9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
import scipy.stats as stats

def confidence_interval_mean(sample_mean, sample_std, sample_size, confidence_level=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Args:
        sample_mean (float): The mean of the sample.
        sample_std (float): The standard deviation of the sample.
        sample_size (int): The size of the sample.
        confidence_level (float, optional): The desired confidence level (e.g., 0.95 for 95%).
                                           Defaults to 0.95.

    Returns:
        tuple: A tuple containing the lower and upper bounds of the confidence interval.
    """
    alpha = 1 - confidence_level
    degrees_of_freedom = sample_size - 1

    # Use t-distribution if sample size is small or population std is unknown (more common)
    if sample_size < 30:
        critical_value = stats.t.ppf(1 - alpha / 2, degrees_of_freedom)
        standard_error = sample_std / np.sqrt(sample_size)
    else:
        critical_value = stats.norm.ppf(1 - alpha / 2)
        standard_error = sample_std / np.sqrt(sample_size)

    margin_of_error = critical_value * standard_error
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
sample_mean = 65
sample_std = 12
sample_size = 50
confidence_level = 0.95

lower_ci, upper_ci = confidence_interval_mean(sample_mean, sample_std, sample_size, confidence_level)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"{confidence_level*100:.0f}% Confidence Interval: ({lower_ci:.2f}, {upper_ci:.2f})")

print("\nSignificance of Confidence Interval:")
print(f"The {confidence_level*100:.0f}% confidence interval ({lower_ci:.2f}, {upper_ci:.2f}) provides a range of values")
print(f"within which we are {confidence_level*100:.0f}% confident that the true population mean lies.")
print("It indicates the uncertainty associated with our estimate of the population mean based on the sample data.")
print("A wider interval suggests greater uncertainty, often due to smaller sample size or higher variability.")
print("If we were to take many random samples and calculate a confidence interval for each, approximately")
print(f"{confidence_level*100:.0f}% of these intervals would contain the true population mean.")
print("The confidence interval helps in making inferences about the population mean without needing to examine the entire population.")
```

```python
# Q11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
import numpy as np

def bayes_theorem(prior, likelihood, evidence):
    """
    Calculates the posterior probability using Bayes' Theorem.

    Args:
        prior (float): The prior probability P(H).
        likelihood (float): The likelihood P(D|H).
        evidence (float): The probability of the evidence P(D).

    Returns:
        float: The posterior probability P(H|D).
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example scenario:
# Suppose we have a medical test for a rare disease.
# Let H be the hypothesis that a person has the disease.
# Let D be the event that the test is positive.

# Given probabilities:
prior_prob_disease = 0.01  # P(H) - Prior probability of having the disease (1%)
likelihood_positive_given_disease = 0.95  # P(D|H) - Probability of a positive test given the person has the disease (True Positive Rate)
prob_positive_given_no_disease = 0.05  # P(D|~H) - Probability of a positive test given the person does not have the disease (False Positive Rate)

# Calculate the evidence P(D) using the law of total probability:
# P(D) = P(D|H) * P(H) + P(D|~H) * P(~H)
prior_prob_no_disease = 1 - prior_prob_disease
evidence_prob_positive = (likelihood_positive_given_disease * prior_prob_disease) + (prob_positive_given_no_disease * prior_prob_no_disease)

# Apply Bayes' Theorem to calculate the posterior probability:
posterior_prob_disease_given_positive = bayes_theorem(prior_prob_disease, likelihood_positive_given_disease, evidence_prob_positive)

print(f"Prior probability of having the disease (P(H)): {prior_prob_disease:.4f}")
print(f"Likelihood of a positive test given disease (P(D|H)): {likelihood_positive_given_disease:.4f}")
print(f"Probability of evidence (positive test) (P(D)): {evidence_prob_positive:.4f}")
print(f"Posterior probability of having the disease given a positive test (P(H|D)): {posterior_prob_disease_given_positive:.4f}")

print("\nExplanation of the Bayesian Inference Process:")
print("1. Prior Probability (P(H)): This is our initial belief about the probability of the hypothesis (having the disease) being true before observing any data (test results). In this case, it's the prevalence of the disease in the population.")
print("2. Likelihood (P(D|H)): This is the probability of observing the data (a positive test) given that the hypothesis is true (the person has the disease). It reflects the test's sensitivity.")
print("3. Evidence (P(D)): This is the overall probability of observing the data (a positive test). It's calculated by considering both the cases where the hypothesis is true and where it's false.")
print("4. Bayes' Theorem: This theorem combines the prior belief with the evidence to update our belief and obtain the posterior probability.")
print("5. Posterior Probability (P(H|D)): This is the updated probability of the hypothesis being true after observing the data. In our example, it's the probability of actually having the disease given that the test result is positive.")
print("\nIn this example, even though the test has a high true positive rate (95%), the posterior probability of having the disease given a positive test is much lower than that. This is because the disease is rare (low prior probability), and there's a non-negligible false positive rate (5%), which contributes to the overall probability of a positive test.")
```

```python
# Q12. Perform a Chi-square test for independence between two categorical variables in Python
import scipy.stats as stats
import pandas as pd

def chi2_test_independence(observed, alpha=0.05):
    """
    Performs a Chi-square test for independence between two categorical variables.

    Args:
        observed (pandas.DataFrame or array-like): The observed contingency table.
                                                  Rows and columns represent the two categorical variables.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the chi2 statistic, p-value, degrees of freedom, and interpretation.
    """
    chi2_statistic, p_value, dof, expected = stats.chi2_contingency(observed)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. There is statistically significant evidence " \
                         f"of an association between the two categorical variables."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence of an association between the two categorical variables."

    return chi2_statistic, p_value, dof, expected, interpretation

# Example usage:
# Create an observed contingency table
observed_data = pd.DataFrame({
    'Smoker': ['Yes', 'Yes', 'No', 'No'],
    'Lung Cancer': ['Yes', 'No', 'Yes', 'No'],
    'Count': [60, 15, 5, 20]
})

contingency_table = pd.pivot_table(observed_data, values='Count', index='Smoker', columns='Lung Cancer')

print("Observed Contingency Table:")
print(contingency_table)

chi2_stat, p_val, degrees_freedom, expected_freq, result_interpretation = chi2_test_independence(contingency_table)

print(f"\nChi-square Statistic: {chi2_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Degrees of Freedom: {degrees_freedom}")
print("\nExpected Frequencies:")
print(expected_freq)
print(f"\nInterpretation: {result_interpretation}")
```

```python
# Q13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculates the expected frequencies for a Chi-square test based on observed data.

    Args:
        observed (pandas.DataFrame or array-like): The observed contingency table.
                                                  Rows and columns represent the two categorical variables.

    Returns:
        numpy.ndarray: The array of expected frequencies.
    """
    observed_array = np.array(observed)
    row_totals = observed_array.sum(axis=1, keepdims=True)
    col_totals = observed_array.sum(axis=0, keepdims=True)
    grand_total = np.sum(observed_array)

    expected_frequencies = (row_totals * col_totals) / grand_total
    return expected_frequencies

# Example usage:
# Create an observed contingency table (same as previous example)
observed_data = pd.DataFrame({
    'Smoker': ['Yes', 'Yes', 'No', 'No'],
    'Lung Cancer': ['Yes', 'No', 'Yes', 'No'],
    'Count': [60, 15, 5, 20]
})

contingency_table = pd.pivot_table(observed_data, values='Count', index='Smoker', columns='Lung Cancer')

print("Observed Contingency Table:")
print(contingency_table)

expected_frequencies = calculate_expected_frequencies(contingency_table)

print("\nExpected Frequencies:")
print(pd.DataFrame(expected_frequencies, index=contingency_table.index, columns=contingency_table.columns))
```

```python
# Q14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution
import scipy.stats as stats
import numpy as np

def goodness_of_fit_test(observed_frequencies, expected_probabilities, alpha=0.05):
    """
    Performs a goodness-of-fit test using the Chi-square statistic.

    Args:
        observed_frequencies (array-like): The observed frequencies for each category.
        expected_probabilities (array-like): The expected probabilities for each category (must sum to 1).
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the chi2 statistic, p-value, and interpretation.
    """
    if not np.isclose(sum(expected_probabilities), 1.0):
        raise ValueError("Expected probabilities must sum to 1.")

    expected_frequencies = np.array(expected_probabilities) * np.sum(observed_frequencies)
    chi2_statistic, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)
    degrees_of_freedom = len(observed_frequencies) - 1

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. The observed frequencies are significantly different " \
                         f"from the expected distribution."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically significant " \
                         f"evidence to suggest that the observed frequencies are different from the expected distribution."

    return chi2_statistic, p_value, degrees_of_freedom, interpretation

# Example usage:
observed = np.array([45, 55, 48, 52]) # Observed frequencies for four categories
expected_proportions = np.array([0.25, 0.25, 0.25, 0.25]) # Expected equal proportions
alpha = 0.05

chi2_stat, p_val, dof, result_interpretation = goodness_of_fit_test(observed, expected_proportions, alpha)

print("Observed Frequencies:", observed)
print("Expected Proportions:", expected_proportions)
print(f"\nChi-square Statistic: {chi2_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"\nInterpretation: {result_interpretation}")

# Example with different expected proportions:
observed_colors = np.array([30, 40, 25, 5]) # Observed frequencies for four colors
expected_probs_colors = np.array([0.3, 0.3, 0.2, 0.2]) # Expected proportions for the colors
alpha_colors = 0.05

chi2_stat_colors, p_val_colors, dof_colors, result_interpretation_colors = goodness_of_fit_test(observed_colors, expected_probs_colors, alpha_colors)

print("\nObserved Frequencies (Colors):", observed_colors)
print("Expected Proportions (Colors):", expected_probs_colors)
print(f"\nChi-square Statistic: {chi2_stat_colors:.3f}")
print(f"P-Value: {p_val_colors:.3f}")
print(f"Degrees of Freedom: {dof_colors}")
print(f"\nInterpretation: {result_interpretation_colors}")
```

```python
# Q16. Implement an F-test using Python to compare the variances of two random samples
import scipy.stats as stats
import numpy as np

def f_test_variances(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two random samples.

    Args:
        sample1 (array-like): The data from the first sample.
        sample2 (array-like): The data from the second sample.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the F-statistic, p-value, and interpretation.
    """
    var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased estimator)
    var2 = np.var(sample2, ddof=1)
    f_statistic = var1 / var2

    dfn = len(sample1) - 1  # Degrees of freedom for numerator
    dfd = len(sample2) - 1  # Degrees of freedom for denominator
    p_value = 2 * min(stats.f.cdf(f_statistic, dfn, dfd), 1 - stats.f.cdf(f_statistic, dfn, dfd)) # Two-tailed test

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. There is statistically significant evidence " \
                         f"that the variances of the two populations are different."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence that the variances of the two populations are different."

    return f_statistic, p_value, dfn, dfd, interpretation

# Example usage:
np.random.seed(42)
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 15, 35)
alpha = 0.05

f_stat, p_val, df_num, df_den, result_interpretation = f_test_variances(group1, group2, alpha)

print(f"Sample 1 Variance: {np.var(group1, ddof=1):.2f}")
print(f"Sample 2 Variance: {np.var(group2, ddof=1):.2f}")
print(f"F-Statistic: {f_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Degrees of Freedom (Numerator): {df_num}")
print(f"Degrees of Freedom (Denominator): {df_den}")
print(f"Interpretation: {result_interpretation}")
```

```python
# Q17. Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results
import scipy.stats as stats
import pandas as pd
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

def perform_anova_and_interpret(data, formula, alpha=0.05):
    """
    Performs a one-way ANOVA test and interprets the results.

    Args:
        data (pandas.DataFrame): The data containing the groups and the measurement variable.
        formula (str): A string representing the formula for the ANOVA model (e.g., 'measurement ~ group').
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the ANOVA table and the interpretation.
    """
    model = ols(formula, data=data).fit()
    anova_table = anova_lm(model)
    p_value = anova_table['PR(>F)'][0]

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. There is statistically significant evidence " \
                         f"that the means of at least one of the groups are different."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence to conclude that the means of the groups are different."

    return anova_table, interpretation

# Example usage:
# Create a sample DataFrame
data = pd.DataFrame({
    'Treatment': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Response': [25, 30, 28, 35, 40, 38, 22, 26, 24]
})

formula = 'Response ~ Treatment'
anova_result, interpretation = perform_anova_and_interpret(data, formula)

print("ANOVA Table:")
print(anova_result)
print("\nInterpretation:", interpretation)
```

```python
# Q19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from statsmodels.stats.api import OLSResults

def check_anova_assumptions(data, measurement_col, group_col):
    """
    Checks the assumptions of ANOVA (normality of residuals, independence, and equal variance).

    Args:
        data (pandas.DataFrame): The data containing the groups and the measurement variable.
        measurement_col (str): The name of the column containing the measurements.
        group_col (str): The name of the column containing the group labels.

    Returns:
        None: Prints the results of the assumption checks and displays plots.
    """
    # Fit the OLS model to get residuals
    formula = f'{measurement_col} ~ C({group_col})'
    model = ols(formula, data=data).fit()
    residuals = model.resid

    print("ANOVA Assumption Checks:")

    # 1. Normality of Residuals
    print("\n1. Normality of Residuals:")
    # Shapiro-Wilk test
    shapiro_test = stats.shapiro(residuals)
    print(f"   Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.3f}, P-value={shapiro_test.pvalue:.3f}")
    alpha = 0.05
    if shapiro_test.pvalue < alpha:
        print(f"   P-value is less than {alpha}, indicating that the residuals may not be normally distributed.")
    else:
        print(f"   P-value is greater than {alpha}, suggesting that the residuals are likely normally distributed.")

    # Histogram and Q-Q plot of residuals
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(residuals, kde=True)
    plt.title('Histogram of Residuals')

    plt.subplot(1, 2, 2)
    stats.probplot(residuals, plot=plt)
    plt.title('Q-Q Plot of Residuals')
    plt.tight_layout()
    plt.show()

    # 2. Independence of Observations
    print("\n2. Independence of Observations:")
    print("   This assumption is typically assessed based on the study design and data collection process.")
    print("   ANOVA assumes that the observations within and between groups are independent of each other.")
    print("   We do not have a statistical test implemented here to directly verify independence. Consider the data collection method.")

    # 3. Equal Variance (Homoscedasticity)
    print("\n3. Equal Variance (Homoscedasticity):")
    # Levene's test
    groups = data[group_col].unique()
    group_data = [data[measurement_col][data[group_col] == g] for g in groups]
    levene_test = stats.levene(*group_data)
    print(f"   Levene's Test: Statistic={levene_test.statistic:.3f}, P-value={levene_test.pvalue:.3f}")
    if levene_test.pvalue < alpha:
        print(f"   P-value is less than {alpha}, indicating that the variances of the groups may not be equal.")
    else:
        print(f"   P-value is greater than {alpha}, suggesting that the variances of the groups are likely equal.")

    # Boxplots to visually inspect variances
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=group_col, y=measurement_col, data=data)
    plt.title('Boxplots of Data by Group (Visual Check for Equal Variance)')
    plt.xlabel(group_col)
    plt.ylabel(measurement_col)
    plt.grid(True)
    plt.show()

# Example usage:
np.random.seed(42)
data_assumptions = pd.DataFrame({
    'Group': ['X'] * 20 + ['Y'] * 20 + ['Z'] * 20,
    'Value': np.concatenate([np.random.normal(5, 1, 20),
                             np.random.normal(6, 1.2, 20),
                             np.random.normal(4.5, 0.9, 20)])
})

check_anova_assumptions(data_assumptions, 'Value', 'Group')

# Example with unequal variances:
data_unequal_var = pd.DataFrame({
    'Group': ['A'] * 20 + ['B'] * 20 + ['C'] * 20,
    'Score': np.concatenate([np.random.normal(10, 1, 20),
                             np.random.normal(12, 2, 20),
                             np.random.normal(9, 0.5, 20)])
})

check_anova_assumptions(data_unequal_var, 'Score', 'Group')

# Example with non-normal residuals:
data_non_normal = pd.DataFrame({
    'Category': ['P'] * 20 + ['Q'] * 20,
    'Measure': np.concatenate([np.random.exponential(scale=2, size=20),
                                np.random.exponential(scale=2.5, size=20)])
})

check_anova_assumptions(data_non_normal, 'Measure', 'Category')
```

```python
# Q20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.anova import anova_lm

def two_way_anova_and_visualize(data, formula, factor1_col, factor2_col, response_col, alpha=0.05):
    """
    Performs a two-way ANOVA test and visualizes the interaction effects.

    Args:
        data (pandas.DataFrame): The data containing the two factors and the response variable.
        formula (str): A string representing the formula for the two-way ANOVA model
                       (e.g., 'response ~ factor1 * factor2').
        factor1_col (str): The name of the first factor column.
        factor2_col (str): The name of the second factor column.
        response_col (str): The name of the response variable column.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the ANOVA table and the interpretation of main and interaction effects.
    """
    model = smf.ols(formula, data=data).fit()
    anova_table = anova_lm(model)

    interpretation = "Two-Way ANOVA Results:\n"
    for effect, row in anova_table.iterrows():
        p_value = row['PR(>F)']
        if p_value < alpha:
            interpretation += f"The effect of '{effect}' is statistically significant (p={p_value:.3f} < {alpha}).\n"
        else:
            interpretation += f"The effect of '{effect}' is not statistically significant (p={p_value:.3f} >= {alpha}).\n"

    # Visualization of interaction effects using interaction plot
    plt.figure(figsize=(8, 6))
    sns.pointplot(x=factor1_col, y=response_col, hue=factor2_col, data=data, capsize=.1)
    plt.title(f'Interaction Plot: {response_col} vs {factor1_col} by {factor2_col}')
    plt.xlabel(factor1_col)
    plt.ylabel(response_col)
    plt.grid(True)
    plt.show()

    # Visualization using boxplots
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=factor1_col, y=response_col, hue=factor2_col, data=data)
    plt.title(f'Boxplot: {response_col} vs {factor1_col} by {factor2_col}')
    plt.xlabel(factor1_col)
    plt.ylabel(response_col)
    plt.legend(title=factor2_col)
    plt.grid(True)
    plt.show()

    return anova_table, interpretation

# Example usage:
np.random.seed(42)
data_two_way = pd.DataFrame({
    'Fertilizer': ['Type A'] * 30 + ['Type B'] * 30,
    'Watering': (['Low'] * 15 + ['High'] * 15) * 2,
    'Yield': np.concatenate([np.random.normal(40, 5, 15),
                             np.random.normal(45, 6, 15),
                             np.random.normal(35, 4, 15),
                             np.random.normal(42, 5, 15)])
})

formula_two_way = 'Yield ~ C(Fertilizer) * C(Watering)'
factor1 = 'Fertilizer'
factor2 = 'Watering'
response = 'Yield'
alpha_level = 0.05

anova_result_two_way, interpretation_two_way = two_way_anova_and_visualize(data_two_way, formula_two_way, factor1, factor2, response, alpha_level)

print("Two-Way ANOVA Table:")
print(anova_result_two_way)
print("\nInterpretation:\n", interpretation_two_way)
```

```python
# Q23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means
import numpy as np
from scipy import stats

def hypothesis_test_normal_mean(population_mean_true, population_std, sample_size, null_hypothesis_mean, alpha=0.05, test_type='two-sided'):
    """
    Simulates random data from a normal distribution and performs a one-sample Z-test
    to evaluate if the sample mean is significantly different from a null hypothesis mean.

    Args:
        population_mean_true (float): The true mean of the population from which data is sampled.
        population_std (float): The standard deviation of the population.
        sample_size (int): The size of the random sample to simulate.
        null_hypothesis_mean (float): The mean stated in the null hypothesis.
        alpha (float, optional): The significance level. Defaults to 0.05.
        test_type (str, optional): The type of hypothesis test ('two-sided', 'less', 'greater').
                                   Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the sample mean, Z-statistic, p-value, and interpretation.
    """
    # Simulate random data from a normal distribution
    sample_data = np.random.normal(population_mean_true, population_std, sample_size)
    sample_mean = np.mean(sample_data)
    standard_error = population_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - null_hypothesis_mean) / standard_error
    p_value = stats.norm.sf(np.abs(z_statistic))

    if test_type == 'two-sided':
        p_value *= 2
        if p_value < alpha:
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence that the population mean is different from {null_hypothesis_mean}."
        else:
            interpretation = f"Fail to reject the null hypothesis. There is not enough statistically significant evidence that the population mean is different from {null_hypothesis_mean}."
    elif test_type == 'less':
        p_value = stats.norm.cdf(z_statistic)
        if p_value < alpha:
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence that the population mean is less than {null_hypothesis_mean}."
        else:
            interpretation = f"Fail to reject the null hypothesis. There is not enough statistically significant evidence that the population mean is less than {null_hypothesis_mean}."
    elif test_type == 'greater':
        if p_value < alpha:
            interpretation = f"Reject the null hypothesis. There is statistically significant evidence that the population mean is greater than {null_hypothesis_mean}."
        else:
            interpretation = f"Fail to reject the null hypothesis. There is not enough statistically significant evidence that the population mean is greater than {null_hypothesis_mean}."
    else:
        raise ValueError("Invalid test_type. Choose 'two-sided', 'less', or 'greater'.")

    return sample_mean, z_statistic, p_value, interpretation

# Example usage:
true_mean = 50
std_dev = 10
n_size = 100
null_mean = 52
significance_level = 0.05
test = 'two-sided'

sample_m, z_stat, p_val, result = hypothesis_test_normal_mean(true_mean, std_dev, n_size, null_mean, significance_level, test)

print(f"Simulated Sample Mean: {sample_m:.2f}")
print(f"Z-Statistic: {z_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(result)

print("\n--- One-tailed test (greater) ---")
sample_m_greater, z_stat_greater, p_val_greater, result_greater = hypothesis_test_normal_mean(true_mean, std_dev, n_size, null_mean, significance_level, 'greater')
print(f"Simulated Sample Mean: {sample_m_greater:.2f}")
print(f"Z-Statistic: {z_stat_greater:.3f}")
print(f"P-Value: {p_val_greater:.3f}")
print(result_greater)
```

```python
# Q24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results
import numpy as np
import scipy.stats as stats

def chi2_test_population_variance(sample_data, hypothesized_variance, alpha=0.05):
    """
    Performs a Chi-square test for the population variance.

    Args:
        sample_data (array-like): The data from the sample.
        hypothesized_variance (float): The hypothesized variance of the population (under the null hypothesis).
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the Chi-square statistic, p-value, degrees of freedom, and interpretation.
    """
    sample_variance = np.var(sample_data, ddof=1)  # Unbiased sample variance
    n = len(sample_data)
    degrees_of_freedom = n - 1
    chi2_statistic = (n - 1) * sample_variance / hypothesized_variance
    p_value = stats.chi2.sf(chi2_statistic, degrees_of_freedom) # Right-tailed test by default

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. There is statistically significant evidence " \
                         f"that the population variance is greater than {hypothesized_variance:.2f}."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence that the population variance is greater than {hypothesized_variance:.2f}."

    return chi2_statistic, p_value, degrees_of_freedom, interpretation

# Example usage:
np.random.seed(42)
population_variance_true = 25
sample = np.random.normal(0, np.sqrt(population_variance_true), 50)
hypothesized_var = 20
alpha_level = 0.05

chi2_stat, p_val, dof, result_interpretation = chi2_test_population_variance(sample, hypothesized_var, alpha_level)

print(f"Sample Variance: {np.var(sample, ddof=1):.2f}")
print(f"Hypothesized Variance: {hypothesized_var:.2f}")
print(f"Chi-square Statistic: {chi2_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Degrees of Freedom: {dof}")
print(f"Interpretation: {result_interpretation}")

# Example for a two-tailed test (adjust p-value and interpretation)
def chi2_test_population_variance_two_tailed(sample_data, hypothesized_variance, alpha=0.05):
    sample_variance = np.var(sample_data, ddof=1)
    n = len(sample_data)
    degrees_of_freedom = n - 1
    chi2_statistic = (n - 1) * sample_variance / hypothesized_variance
    p_value_right = stats.chi2.sf(chi2_statistic, degrees_of_freedom)
    p_value_left = stats.chi2.cdf(chi2_statistic, degrees_of_freedom)
    p_value_two_tailed = 2 * min(p_value_left, p_value_right)

    if p_value_two_tailed < alpha:
        interpretation = f"The p-value ({p_value_two_tailed:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. There is statistically significant evidence " \
                         f"that the population variance is different from {hypothesized_variance:.2f}."
    else:
        interpretation = f"The p-value ({p_value_two_tailed:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence that the population variance is different from {hypothesized_variance:.2f}."

    return chi2_statistic, p_value_two_tailed, degrees_of_freedom, interpretation

print("\n--- Two-tailed test for population variance ---")
chi2_stat_two, p_val_two, dof_two, result_two = chi2_test_population_variance_two_tailed(sample, hypothesized_var, alpha_level)
print(f"Sample Variance: {np.var(sample, ddof=1):.2f}")
print(f"Hypothesized Variance: {hypothesized_var:.2f}")
print(f"Chi-square Statistic: {chi2_stat_two:.3f}")
print(f"P-Value (Two-tailed): {p_val_two:.3f}")
print(f"Degrees of Freedom: {dof_two}")
print(f"Interpretation: {result_two}")
```

```python
# Q25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.proportion import proportions_ztest

def two_sample_proportion_z_test(successes1, n1, successes2, n2, alpha=0.05, alternative='two-sided'):
    """
    Performs a two-sample Z-test for comparing proportions.

    Args:
        successes1 (int): Number of successes in the first sample.
        n1 (int): Total number of observations in the first sample.
        successes2 (int): Number of successes in the second sample.
        n2 (int): Total number of observations in the second sample.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The alternative hypothesis ('two-sided', 'smaller', 'larger').
                                     Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the Z-statistic, p-value, and interpretation.
    """
    count = np.array([successes1, successes2])
    nobs = np.array([n1, n2])

    z_statistic, p_value = proportions_ztest(count, nobs, alternative=alternative)

    if p_value < alpha:
        if alternative == 'two-sided':
            interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                             f"We reject the null hypothesis. There is statistically significant evidence " \
                             f"that the proportions in the two groups are different."
        elif alternative == 'smaller':
            interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                             f"We reject the null hypothesis. There is statistically significant evidence " \
                             f"that the proportion in the first group is smaller than the proportion in the second group."
        elif alternative == 'larger':
            interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                             f"We reject the null hypothesis. There is statistically significant evidence " \
                             f"that the proportion in the first group is larger than the proportion in the second group."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence to conclude that the proportions in the two groups are different."

    return z_statistic, p_value, interpretation

# Example usage:
successes_group1 = 55
n_group1 = 100
successes_group2 = 40
n_group2 = 90
alpha_level = 0.05
alternative_hypothesis = 'two-sided'

z_stat, p_val, result_interpretation = two_sample_proportion_z_test(
    successes_group1, n_group1, successes_group2, n_group2, alpha_level, alternative_hypothesis
)

print(f"Group 1: Successes = {successes_group1}, Total = {n_group1}, Proportion = {successes_group1/n_group1:.3f}")
print(f"Group 2: Successes = {successes_group2}, Total = {n_group2}, Proportion = {successes_group2/n_group2:.3f}")
print(f"Z-Statistic: {z_stat:.3f}")
print(f"P-Value: {p_val:.3f}")
print(f"Interpretation: {result_interpretation}")

print("\n--- One-tailed test (larger) ---")
z_stat_larger, p_val_larger, result_larger = two_sample_proportion_z_test(
    successes_group1, n_group1, successes_group2, n_group2, alpha_level, 'larger'
)
print(f"Z-Statistic: {z_stat_larger:.3f}")
print(f"P-Value: {p_val_larger:.3f}")
print(f"Interpretation: {result_larger}")
```

```python
# Q26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def f_test_variances_visualize(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two random samples,
    interprets the results, and visualizes the distributions.

    Args:
        sample1 (array-like): The data from the first sample.
        sample2 (array-like): The data from the second sample.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the F-statistic, p-value, degrees of freedom, and interpretation.
    """
    var1 = np.var(sample1, ddof=1)  # Sample variance (unbiased estimator)
    var2 = np.var(sample2, ddof=1)
    f_statistic = var1 / var2

    dfn = len(sample1) - 1  # Degrees of freedom for numerator
    dfd = len(sample2) - 1  # Degrees of freedom for denominator
    p_value = 2 * min(stats.f.cdf(f_statistic, dfn, dfd), 1 - stats.f.cdf(f_statistic, dfn, dfd)) # Two-tailed test

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. There is statistically significant evidence " \
                         f"that the variances of the two populations are different."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically " \
                         f"significant evidence that the variances of the two populations are different."

    # Visualization
    plt.figure(figsize=(10, 6))
    sns.histplot(sample1, kde=True, color='skyblue', label='Sample 1')
    sns.histplot(sample2, kde=True, color='salmon', label='Sample 2')
    plt.title('Distribution of Two Samples')
    plt.xlabel('Data Values')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Sample 1 Variance: {var1:.2f}")
    print(f"Sample 2 Variance: {var2:.2f}")
    print(f"F-Statistic: {f_statistic:.3f}")
    print(f"P-Value: {p_value:.3f}")
    print(f"Degrees of Freedom (Numerator): {dfn}")
    print(f"Degrees of Freedom (Denominator): {dfd}")
    print(f"Interpretation: {interpretation}")

    return f_statistic, p_value, dfn, dfd

# Example usage:
np.random.seed(42)
group_a = np.random.normal(50, 10, 30)
group_b = np.random.normal(55, 15, 35)
alpha_level = 0.05

f_stat, p_val, df_num, df_den = f_test_variances_visualize(group_a, group_b, alpha_level)
```

```python
# Q27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results
import numpy as np
import scipy.stats as stats

def chi2_goodness_of_fit_simulated(expected_probabilities, num_observations=100, alpha=0.05):
    """
    Simulates categorical data based on expected probabilities and performs a
    Chi-square goodness-of-fit test.

    Args:
        expected_probabilities (array-like): The expected probabilities for each category (must sum to 1).
        num_observations (int, optional): The number of observations to simulate. Defaults to 100.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing the observed frequencies, expected frequencies,
               Chi-square statistic, p-value, degrees of freedom, and interpretation.
    """
    if not np.isclose(sum(expected_probabilities), 1.0):
        raise ValueError("Expected probabilities must sum to 1.")

    num_categories = len(expected_probabilities)
    simulated_data = np.random.choice(range(num_categories), size=num_observations, p=expected_probabilities)
    observed_frequencies = np.bincount(simulated_data, minlength=num_categories)
    expected_frequencies = np.array(expected_probabilities) * num_observations
    degrees_of_freedom = num_categories - 1

    chi2_statistic, p_value = stats.chisquare(observed_frequencies, f_exp=expected_frequencies)

    if p_value < alpha:
        interpretation = f"The p-value ({p_value:.3f}) is less than the significance level ({alpha}). " \
                         f"We reject the null hypothesis. The observed frequencies are significantly different " \
                         f"from the expected distribution."
    else:
        interpretation = f"The p-value ({p_value:.3f}) is greater than the significance level ({alpha}). " \
                         f"We fail to reject the null hypothesis. There is not enough statistically significant " \
                         f"evidence to suggest that the observed frequencies are different from the expected distribution."

    print("Simulated Data Analysis:")
    print(f"Number of Observations: {num_observations}")
    print("Expected Probabilities:", expected_probabilities)
    print("Observed Frequencies:", observed_frequencies)
    print("Expected Frequencies:", expected_frequencies)
    print(f"Chi-square Statistic: {chi2_statistic:.3f}")
    print(f"P-Value: {p_value:.3f}")
    print(f"Degrees of Freedom: {degrees_of_freedom}")
    print(f"Interpretation: {interpretation}")

    return observed_frequencies, expected_frequencies, chi2_statistic, p_value, degrees_of_freedom, interpretation

# Example usage:
expected_probs = np.array([0.4, 0.3, 0.2, 0.1])
num_obs = 150
alpha_level = 0.05

observed_freq, expected_freq, chi2_stat, p_val, dof, result = chi2_goodness_of_fit_simulated(expected_probs, num_obs, alpha_level)
```
